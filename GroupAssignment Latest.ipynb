```python
import tkinter as tk
from tkinter import messagebox

class TicTacToe:
    def __init__(self):
        self.window = tk.Tk()
        self.window.title("Group 8 Project - Tic Tac Toe")  # Set title
        self.window.config(bg="lightblue")  # Set window background to light blue
        self.current_player = "X"
        self.board = [""] * 9
        self.buttons = []

        self.Create_Window()
        self.window.mainloop()

    def Create_Window(self):
        for i in range(3):
            for j in range(3):
                btn = tk.Button(self.window, text="", font=("Arial", 25, "bold"), height=3, width=6,
                                command=lambda idx=i*3+j: self.on_click(idx), bg="lightblue")  # Increase size of grid buttons
                btn.grid(row=i, column=j, padx=5, pady=5)  # Add some padding for better spacing
                self.buttons.append(btn)

        self.reset_button = tk.Button(self.window, text="Reset", font=("Arial", 12, "bold"),
                                      command=self.reset_board, bg="lightblue", height=2, width=12)  # Smaller Reset button
        self.reset_button.grid(row=3, column=0, columnspan=3, padx=5, pady=5)

    def on_click(self, idx):
        if self.board[idx] == "" and not self.check_winner():
            self.board[idx] = self.current_player
            self.buttons[idx].config(text=self.current_player, fg="darkblue")  # Set text color to dark blue

            if self.check_winner():
                messagebox.showinfo("Game Over", f"Player {self.current_player} wins!")
                return

            if "" not in self.board:
                messagebox.showinfo("Game Over", "It's a tie!")
                return

            self.current_player = "O" if self.current_player == "X" else "X"

    def check_winner(self):
        win_conditions = [(0, 1, 2), (3, 4, 5), (6, 7, 8),
                          (0, 3, 6), (1, 4, 7), (2, 5, 8),
                          (0, 4, 8), (2, 4, 6)]
        for a, b, c in win_conditions:
            if self.board[a] == self.board[b] == self.board[c] and self.board[a] != "":
                # Highlight winning buttons with green
                for btn in (self.buttons[a], self.buttons[b], self.buttons[c]):
                    btn.config(bg="lightgreen")  # Highlight winning cells in green
                self.window.after(1000, self.reset_to_lightblue)  # After 1 second, reset to light blue
                return True
        return False

    def reset_to_lightblue(self):
        for btn in self.buttons:
            btn.config(bg="lightblue")  # Reset the background color of the buttons to light blue after the game ends

    def reset_board(self):
        self.board = [""] * 9
        self.current_player = "X"
        for btn in self.buttons:
            btn.config(text="", bg="lightblue")  # Reset the buttons' background color to light blue

# Run the game
TicTacToe()
```
